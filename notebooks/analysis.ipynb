```python
sql("""
SELECT
    t.txn_id,
    t.txn_ts,
    t.amount_uzs,
    t.status          AS txn_status,
    t.auth_response,
    t.card_scheme,
    t.is_cross_region,

    term.terminal_id,
    term.terminal_type,
    term.region_id     AS terminal_region_id,

    m.merchant_id,
    m.legal_name,
    m.brand_name,
    m.size_band,
    m.status           AS merchant_status,

    mc.category_id,
    mc.category_name,
    mc.risk_tier,

    r.region_id         AS merchant_region_id,
    r.region_name,
    r.macro_zone,

    s.settlement_id,
    s.batch_date,
    s.settled_ts,
    s.gross_amount_uzs,
    s.commission_uzs,
    s.net_amount_uzs,

    d.dispute_id,
    d.reason_code,
    d.dispute_type,
    d.dispute_amount_uzs,
    d.status            AS dispute_status

FROM ds_transactions_1 t
JOIN ds_terminals_1 term        ON t.terminal_id   = term.terminal_id
JOIN ds_merchants_1 m           ON term.merchant_id = m.merchant_id
JOIN ds_merchant_categories mc  ON m.category_id   = mc.category_id
JOIN ds_regions r                ON m.region_id     = r.region_id
LEFT JOIN ds_settlements s      ON t.settlement_id = s.settlement_id
LEFT JOIN ds_disputes d          ON t.txn_id        = d.txn_id;
""")
```
